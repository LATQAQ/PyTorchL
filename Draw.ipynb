```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('data3.csv')
```

```python
for i in range(1,92):
    data.age1[i-1] = str(i)
    data.age2[i-1] = str(i)
    data.age3[i-1] = str(i)
    data.age4[i-1] = str(i)
    data.age5[i-1] = str(i)
data = data.dropna(axis=1)
data
```

```python
import random
from matplotlib.ticker import MaxNLocator
plt.figure(figsize=(15,20))

bar_plot = sns.barplot(y = [i for i in range(1,92)], x = [i*random.uniform(0.97, 1.03) for i in data.population1.values], color=sns.color_palette("Paired")[0] , order=data.age1.values[::-1],orient='h', width=1)
bar_plot = sns.barplot(y = [i for i in range(1,92)], x = [-i*random.uniform(0.97, 1.03) for i in data.population1.values], color=sns.color_palette("ch:2.5,-.2,dark=.3")[0],order=[i for i in range(1,92)][::-1],orient='h', width=1)

# plt.rcParams['font.sans-serif'] = ['SimHei']

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = True
bar_plot.set(xlabel="Population", ylabel="Age", title = "å¹´é¾„ç»“æž„")
plt.show()
```
