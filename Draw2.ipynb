```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_excel('data_new.xlsx',sheet_name='1')
data
```

```python

data_d = np.array([data.AQI,data.PM10,data.O3,data.SO2,data.PM2_5,data.NO2,data.CO,data.V13305,data.V10004_700,data.V11291_700,data.V12001_700,data.V13003_700]).T
(data_d.astype('int64')*10)//10
```

```python
df = pd.DataFrame(data_d.astype('int64'))
df_ = pd.DataFrame(data_d.astype('int64'))
df = df.replace(-9223372036854775808,-10)
df_ = df.replace(-9223372036854775808,-10)
df
```
